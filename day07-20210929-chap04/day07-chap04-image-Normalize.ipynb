```python
import matplotlib.pyplot as plt
bear = plt.imread('bear.jpg')
bird = plt.imread('bird.jpg')
print(bear.shape, bird.shape)
print('bear pixel', bear.flatten().min(), bear.flatten().max())
plt.imshow(bear)
plt.show()
plt.hist(bear.flatten(), edgecolor='red')
plt.show()
```

```python
# 資料正規化（Normalize）


import matplotlib.pyplot as plt
bear_small = plt.imread('bear.jpg') /255.0

print(bear_small.shape)
print('bear pixel / 255.0', bear_small.flatten().min(), bear_small.flatten().max())
plt.imshow(bear_small)
plt.show()
plt.hist(bear_small.flatten(), edgecolor='red')
plt.show()
```

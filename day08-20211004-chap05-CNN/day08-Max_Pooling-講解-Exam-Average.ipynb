```python
%matplotlib inline
import matplotlib.pyplot as plt
from skimage import color
bricks = color.rgb2gray(plt.imread('andy2.jpg'))
plt.imshow(bricks, cmap='gray')
plt.show()
print(bricks.shape)
```

```python
import numpy as np
np.min(bricks[:, :]), np.max(bricks[:, :])
```

```python
pool = (2,2)
max_height= int(bricks.shape[0]//pool[0])
max_width = int(bricks.shape[1]//pool[1])
bricks_max = np.zeros((max_height, max_width))
for ii in range(max_height):
    for jj in range(max_width):
                bricks_max[ii, jj] = np.mean(bricks[ii*2:ii*2+pool[0], jj*2:jj*2+pool[0]])
```

```python
bricks_max.min(), bricks_max.max()
```

```python
plt.imshow(bricks_max, cmap='gray')
plt.show()
print(bricks_max.shape)
```

```python
pool = (15, 15)
max_height= int(bricks.shape[0]//pool[0])
max_width = int(bricks.shape[1]//pool[1])
bricks_max = np.zeros((max_height, max_width))
for ii in range(max_height):
    for jj in range(max_width):
                bricks_max[ii, jj] = np.mean(bricks[ii*pool[0]:ii*pool[0]+pool[0], jj*pool[1]:jj*pool[1]+pool[1]])
            
plt.imshow(bricks_max, cmap='gray')
plt.show()
print(bricks_max.shape)            
```

```python
# 講解
# https://en.wikipedia.org/wiki/Kernel_(image_processing)
# import data
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
from skimage import color
img = color.rgb2gray(plt.imread('andy.jpg'))
plt.imshow(img, cmap='gray')
plt.show()
```

```python
kernel = np.array([[  1,   1],
                   [ -1 , -1]] )

height, width = img.shape[0]-1, img.shape[1]-1
conv = np.zeros((height, width))
for ii in range(height):
    for jj in range(width):
        val = np.sum(kernel * img[ii:ii+2, jj:jj+2])
        conv[ii, jj] = val
plt.imshow(conv, cmap='gray')
plt.show()
```

```python
kernel = np.array([[-1, 1],
                   [-1, 1]] )

height, width = img.shape[0]//2, img.shape[1]//2
conv = np.zeros((height, width))
for ii in range(height):
    for jj in range(width):
        val = np.sum(kernel * img[ii*2:ii*2+2, jj*2:jj*2+2])
        conv[ii, jj] = val
plt.imshow(conv, cmap='gray')
plt.show()
```

```python
kernel = np.array([[-1, -1, -1],
                   [-1,  8, -1],
                   [-1, -1, -1] ] )

height, width = img.shape[0]-3+1, img.shape[1]-3+1
conv = np.zeros((height, width))
for ii in range(height):
    for jj in range(width):
        val = np.sum(kernel * img[ii:ii+3, jj:jj+3])
        conv[ii, jj] = val
plt.imshow(conv, cmap='gray')
plt.show()
```

```python
kernel = np.array([[ 1,  1,  1],
                   [ 1,  1,  1],
                   [ 1,  1,  1] ] ) /9

height, width = img.shape[0]-3+1, img.shape[1]-3+1
conv = np.zeros((height, width))
for ii in range(height):
    for jj in range(width):
        val = np.sum(kernel * img[ii:ii+3, jj:jj+3])
        conv[ii, jj] = val
plt.imshow(conv, cmap='gray')
plt.title('Blur')
plt.show()
plt.imshow(img, cmap='gray')
plt.show()
```
